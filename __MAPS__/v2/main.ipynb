```python
visualize_map(
    gangjin,
    background_path=r"gangjin/pcd_2d.png",
    grid_alpha=Config.alpha,
    show_outside=Config.show_outside,
    show_inside=Config.show_inside,
    show_installable=Config.show_installable,
    show_constraint=Config.show_constraint,
    save_path=r"results/gangjin_map.png",
)
```
