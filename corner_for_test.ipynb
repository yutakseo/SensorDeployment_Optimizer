```python
from OuterDeployment.HarrisCorner import HarrisCorner
from Tools.MapLoader import MapLoader
from Tools.PlotTools import VisualTool
```

```python
vis = VisualTool(show=True, size=(4,3))
```

```python
MAP = MapLoader.load("map_280x250.bot")
vis.showJetMap(MAP)
```

```python
test = HarrisCorner(MAP=MAP)
test.run(map=MAP)
print("Total number of sensors : ", test.total_sensors)
print("Corner positions is : ", test.corner_positions)
```

```python
vis.showJetMap_circle(map_data=MAP,sensor_positions=test.corner_positions)
```

```python
temp = HarrisCorner(MAP=MAP)
```

```python
origin_map = temp.map_data
vis.showJetMap(origin_map)
```

```python
gaussian_blur = temp.gaussianBlur(map=MAP)
vis.showJetMap(gaussian_blur)
```

```python
harris_response_map = temp.harrisCorner(
    map=gaussian_blur,
    block_size=3,
    ksize=3,
    k=0.05,
    threshold=0.2)
vis.showJetMap(harris_response_map)
```

```python
nms_map = temp.non_max_suppression(
    response=harris_response_map,
    nmx_threshold=0.01,
    dilate_size=9
    )
vis.showJetMap(nms_map)
vis.showJetMap(MAP)
```
